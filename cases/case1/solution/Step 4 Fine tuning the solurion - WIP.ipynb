```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline
```

```python
data = 'SC_data_try4.csv'

df = pd.read_csv(data)

df.shape

df.head()
```

```python
print(list(df.columns))
```

```python
df['default_flag'].value_counts()
```

```python
df.isnull().sum()
```

```python
y = df['default_flag']
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 0)
```

```python
#Stratifing the train population to balance the '1' and '0' to better train the model in recognizing the probability to default

df_data=X_train
strat1 = df_data.loc[df_data['default_flag']==1] # Calculate the number of rows to match the desired proportion

temp0 = df_data.loc[df_data['default_flag']==0]
strat0=temp0.sample(frac=0.05, random_state=0)

X_strat = pd.concat([strat1, strat0]) 
y_strat = X_strat['default_flag']
X_strat.shape
```

```python
#check the size of the '0' population
strat0.shape
```

```python
X_train_tmp=X_train.drop(['default_flag'], axis=1)
y_train_tmp = X_train['default_flag']

X_strat = X_strat.drop(['default_flag'], axis=1)
X_test = X_test.drop(['default_flag'], axis=1)
```

```python
X_train=X_strat
y_train=y_strat
```

```python
X_train.shape, X_test.shape
```

```python
# check data types in X_train

X_train.dtypes
```

```python
X_train.head()
```

# Decision Tree Classifier with "Gini"

```python
#preparing to apply the Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index according to the optimized paramethers

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=10, max_features=10, min_samples_split=10, random_state=0)

# max_features=30,
clf_gini.fit(X_train, y_train)
```

```python
#setting a threshold for the cut off/misclassification costs 

y_pred_gini = (clf_gini.predict_proba(X_test)[:,1] >= 0.1).astype(bool)
```

```python
zzz=clf_gini.predict_proba(X_test)[:,1]
plt.hist(x=zzz, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

```python
y_pred_train_gini = (clf_gini.predict_proba(X_train)[:,1] >= 0.1).astype(bool)
y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(clf_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_gini)

print('Confusion matrix\n\n', cm)

# visualize confusion matrix with seaborn heatmap

cm = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu')
```

```python
y_pred_train_tmp = clf_gini.predict(X_train_tmp)

cm_train = confusion_matrix(y_train_tmp, y_pred_train_tmp)

print('Confusion matrix\n\n', cm_train)
```

```python

cm_train = confusion_matrix(y_train, y_pred_train_gini)

print('Confusion matrix\n\n', cm_train)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, accuracy_score, confusion_matrix 
fpr, tpr, thresholds = roc_curve(y_test, y_pred_gini, drop_intermediate=True)

f, ax = plt.subplots(figsize=(9, 6))
_ = plt.plot(fpr, tpr, [0,1], [0, 1])
_ = plt.title('AUC ROC')
_ = plt.xlabel('False positive rate')
_ = plt.ylabel('True positive rate')
plt.style.use('seaborn')

plt.savefig('auc_roc.png', dpi=600)
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(clf_gini.fit(X_train, y_train)) 
```

```python
import graphviz

dot_data = tree.export_graphviz(clf_gini, out_file=None, 
                              feature_names=X_train.columns,  
                              class_names=None,  
                              filled=True, rounded=True,  
                              special_characters=True)



graph = graphviz.Source(dot_data)

graph.render('dtree_render_example_gini',view=True)

```
